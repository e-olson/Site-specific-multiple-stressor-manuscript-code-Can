```python
ivars=['tos','sos','spco2','phos','chlos','o2os','phosC']
```

```python
recalc=False
if recalc:
    # create table
    dfTable=dfInfoBuoy.loc[:,['datasetID','shortTitle','Lat','Lon']].copy(deep=True)
    # create columns
    for iv in ivars:
        dfTable[iv+'_N']=np.nan
        dfTable[iv+'_N2']=np.nan
        dfTable[iv+'_ofit']=False
        dfTable[iv+'_mfit']=False
    dfTable['AlkFit']='None'

    for dsid in dfTable.datasetID:
        print(dsid)
        [lat,lon], ldict, udict, df0, df = OAP.loadOAPBuoy(dfInfoBuoy,dsid,freq='daily',filtSSS=True,cCalcs=True)
        for iv in ivars:
            ovar=bc.vardict[iv]
            dfTable.loc[dfTable.datasetID==dsid,[iv+'_N']]=np.sum(~pd.isnull(df[ovar]))
            # save numver of obs_val in N_2
            # check whether moorcomp obs_targetdeseas has real values and whether obs_targetdeseas_b has real values
            # save under iv_ofit (has obs_targetdeseas) and iv_mfit (has obs_targetdeseas_b)
            ic=bc.loadMoorComp(dsid,iv,'daily',path='/work/ebo/calcs/buoyCompTS/presentday/comps/')
            dfTable.loc[dfTable.datasetID==dsid,[iv+'_N2']]=np.sum(~pd.isnull(ic.obs_val))
            dfTable.loc[dfTable.datasetID==dsid,[iv+'_ofit']]=(np.sum(~pd.isnull(ic.obs_targetdeseas))>0)
            dfTable.loc[dfTable.datasetID==dsid,[iv+'_mfit']]=(np.sum(~pd.isnull(ic.obs_targetdeseas_b))>0)
        stitle=dfInfoBuoy.loc[dfInfoBuoy.datasetID==dsid,['shortTitle']].values[0][0]
        dfTable.loc[dfTable.datasetID==dsid,['AlkFit']]=AF.OAP_Alk_cit(stitle)
    dfTable.to_csv('table.csv')
else: # load from file
    dfTable=pd.read_csv('table.csv',index_col=0)
```

```python
def formatn(n,ofit,mfit):
    if ofit:
        out=r"\textbf{"+str(int(n))+r"}"
    #elif not mfit:
    #    out='-'
    #    #out=r"\textcolor{gray}{"+str(int(n))+r"}"
    else:
        out=str(int(n))
    return out
```

```python
def fmtName(name,spco2_N,AlkFit):
    if spco2_N/365>=thresh_yrs_stressors and AlkFit is not None:
        return name+'$^*$'
    elif name == 'La Parguera':
        return name+'$^*$'
    else:
        return name
```

```python
trdict={'North Atlantic Ocean':'North Atlantic',
        'Southeast Pacific Ocean':'Southeast Pacific',
        'Northeast Pacific Ocean':'Northeast Pacific',
        'Northwest Pacific Ocean':'Northwest Pacific',
        'Central Pacific Ocean': 'Central Pacific',
        'Equatorial Pacific Ocean':'Equatorial Pacific',
        'Pacific island coral reef': 'Pacific Island CR',
        'Caribbean coral reef': 'Caribbean CR',
        'Atlantic coral reef': 'Atlantic CR',
        'Gulf of Mexico coast': 'Gulf Mex. coast',
        'MOSEAN/WHOTS': 'MOSEAN/ WHOTS'}
```

```python
def fmt2(ss):
    if type(ss)==str:
        if ss in trdict.keys():
            return trdict[ss]
        else:
            return ss
    elif type(ss)==float:
        return f'{ss:.1f}'
    else:
        return str(ss)
```

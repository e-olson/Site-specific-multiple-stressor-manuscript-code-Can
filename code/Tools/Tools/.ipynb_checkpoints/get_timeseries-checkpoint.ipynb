```python
indlist=[]
for ind, row in dfb.iterrows():
    if row.modBathy>0:
        fnameout=diagsPP.extractedTSPath(scen,row.datasetID,row.jj,row.ii)
        if not os.path.exists(fnameout):
            cf.mkdirs(fnameout)
            print(fnameout)
            indlist.append(ind)
            flongTi=flongT.isel(j=row.jj,i=row.ii)
            flongSi=flongS.isel(j=row.jj,i=row.ii)
            flongTi.to_netcdf(fnameout,'w')
            flongSi.to_netcdf(fnameout,'a')
            del flongTi 
            del flongSi
```
